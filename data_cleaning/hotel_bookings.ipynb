```python
import pandas as pd

dt = pd.read_csv("hotel_bookings.csv")
#dt.isna().sum()
 
#dt.duplicated().sum()
dt["Arrival_date"] = dt["arrival_date_day_of_month"].astype(str) + "-" + dt["arrival_date_month"] + "-" + dt["arrival_date_year"].astype(str)
dt = dt.drop(columns=[
    "arrival_date_month",
    "arrival_date_day_of_month",
    "arrival_date_year"
])
```

```python
#re-arrange the arrival column 
cols = list(dt.columns)
cols.remove("Arrival_date")
cols.insert(3, "Arrival_date")
dt = dt[cols]

#re-arrange the arrival column 
cols = list(dt.columns)
cols.remove("reservation_status_date")
cols.insert(4, "reservation_status_date")
dt = dt[cols]

#format the datetime format 
dt["Arrival_date"] = pd.to_datetime(
    dt["Arrival_date"],
    dayfirst=True
)
dt
```
